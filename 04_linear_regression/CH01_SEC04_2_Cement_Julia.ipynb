```julia
using Plots, LinearAlgebra, CSV
using DataFrames
```

```julia
# Load dataset
A = CSV.read("/home/jose/julia_projects/estudos_Julia/04_linear_regression/hald_ingredients.csv", Tables.matrix);
b = CSV.read("/home/jose/julia_projects/estudos_Julia/04_linear_regression/hald_heat.csv", Tables.matrix);
```

```julia
u, s, v = svd(A);
```

```julia
x = v * inv(Diagonal(s)) * transpose(u) * b;
```

```julia
plot(b, label="Heat Data") # True relationship
plot!(A * x, line=:dash,label="Regression")
```
