```julia
using Plots, LinearAlgebra
```

```julia
x = 3;
# a = aa = -2:0.25:2
a = range(-2,2,step=0.25) |> collect;
# a = reshape(a,1);
b = x*a + rand((length(a)));
```

```julia
u, s, v = svd(a, full=false);
```

```julia
xtilde = v * inv(Diagonal(s)) * transpose(u) * b; # Least-square fit
```

```julia
plot(a, x*a) # True relationship
scatter!(a, b) # Noisy measurements
plot!(a, xtilde .* a, line=:dash)
```
